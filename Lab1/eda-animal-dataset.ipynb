```python
import os
import cv2
import numpy as np
from imutils import paths
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

# EDA

```python
def count_images(data_path):
    categories = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]
    counts = {category: len(os.listdir(os.path.join(data_path, category))) for category in categories}
    return counts

data_path = '/kaggle/input/animal/Animal_Dataset'
image_counts = count_images(data_path)
print(image_counts)
```

```python
def plot_image_counts(image_counts):
    sns.barplot(x=list(image_counts.keys()), y=list(image_counts.values()))
    plt.title("Number of Images per Category")
    plt.xlabel("Category")
    plt.ylabel("Number of Images")
    plt.show()

plot_image_counts(image_counts)
```
